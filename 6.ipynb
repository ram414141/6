```python
import numpy as np
```

```python
def sinplot(flip=1):
    x=np.linspace(0,14,100)
    for i in range(1,5):
        plt.plot(x,np.sin(x+i*.5)*(7-i)*flip)
        
```

```python
sinplot()
plt.show()
```
